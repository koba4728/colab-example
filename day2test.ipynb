```python
# prompt: box plot

import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create the box plot
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])

# Customize the plot (optional)
plt.title('Box Plot Example')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.grid(True)

# Show the plot
plt.show()
```
